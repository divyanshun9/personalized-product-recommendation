# Item to Item Recommendation ( Content Based filtering)

```python
import numpy as np
import pandas as pd
```

```python
product_df=pd.read_csv('final_final_product_dataset.csv')
product_df
```

```python
product_df['final']=product_df['name']+product_df['main_category']+product_df['sub_category']
product_df
```

```python
df=product_df[['product_id','name','final']]
```

```python
def makelower(obj):
    x=obj.lower()
    return x
```

```python
df['final']=df['final'].apply(makelower)
```

```python
df['final']
```

```python
from nltk.stem.porter import PorterStemmer
p=PorterStemmer()
```

```python
def Stemmer(obj):
    l=[]
    for i in obj.split():
        l.append(p.stem(i))
    return l
```

```python
df['final']=df['final'].apply(Stemmer)
```

```python
df['final'][0]
```

```python
def joinList(obj):
    x=" ".join(obj)
    return x
```

```python
df['final']=df['final'].apply(joinList)
```

```python
df['final']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# CountVectorizer --->  Used to transform a given text into vector on the basis of count of each word
c = CountVectorizer(max_features=5000,stop_words='english')
vector=c.fit_transform(df['final'])
vector
```

```python
v=vector.toarray()
v
```

```python
# cosine_similarity --->  Finds the cosine angle between any two vectors for similarity
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(v)
similarity
```

```python
np.save("sim.npy", similarity)
```

```python
sim = np.load("sim.npy")
```

```python
all_res = np.zeros((1, 50))
```

```python
sim
```

```python
def Idx(product_id):
    for i in range(len(df)):
        if df['product_id'][i]==product_id:
            return i
```

```python
def RecommendProduct(product):
    product_index=Idx(product)
    List=sorted(enumerate(similarity[product_index]),key=lambda x: x[1],reverse=True)
    List=List[1:51]
    product_id_list=[]
    for i in List:
        product_id_list.append(df.loc[i[0]]['product_id'])
    
    return product_id_list
```

```python
RecommendProduct(4588)
```

```python
dic={}
length=len(product_df)
for i in range(length):
    rec_list = RecommendProduct(df.loc[i]['product_id'])
    dic[df.loc[i]['product_id']]=rec_list
    print("saved for:" + str(i))
    
dic
```

```python
import pickle
pickle.dump(dic,open('item_to_item_recommend.pkl','wb'))
```
