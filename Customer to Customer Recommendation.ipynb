# Customer to Customer similarity(Collaborative Based Filtering)

```python
import numpy as np
import pandas as pd
```

```python
product_df=pd.read_csv('final_final_product_dataset.csv')
product_df
```

```python
customer_df=pd.read_csv('final_customer_dataset.csv')
customer_df
```

```python
pt=customer_df.pivot_table(index='customer_id',columns='product_id',values='rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
pt
```

```python
pt.columns
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
```

```python
similarity_score
```

```python
def recommend(customer_id):
    index=np.where(pt.index==customer_id)[0][0]
    simi=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    customer_id_list=[]
    for i in simi:
        customer_id_list.append(pt.index[i[0]])
        
    return customer_id_list
```

```python
recommend(50002)
```

```python
dic={}
for i in range(50000,50050):
    get=recommend(i)
    dic[i]=get
    print("saved for:" + str(i))
dic
```

```python
import pickle
pickle.dump(dic,open('customer_to_customer_recommend.pkl','wb'))
```
