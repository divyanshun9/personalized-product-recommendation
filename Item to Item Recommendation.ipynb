# Age Location Based Recommender Model

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('final_customer_dataset.csv')
df
```

```python
p={"18-20": [] , "20-25": [], "25-30": [], "30-35": [], "35-40": []}
check={"18-20": False , "20-25": False, "25-30": False, "30-35": False, "35-40": False}
for i in range(len(df)):
    if df.loc[i]['age']<=20:
        check["18-20"]=True
        if df.loc[i]['location'] not in p["18-20"]:
            p["18-20"].append(df.loc[i]['location'])
            
    elif df.loc[i]['age']<=25:
        check["20-25"]=True
        if df.loc[i]['location'] not in p["20-25"]:
            p["20-25"].append(df.loc[i]['location'])
            
    elif df.loc[i]['age']<=30:
        check["25-30"]=True
        if df.loc[i]['location'] not in p["25-30"]:
            p["25-30"].append(df.loc[i]['location'])
            
    elif df.loc[i]['age']<=35:
        check["30-35"]=True
        if df.loc[i]['location'] not in p["30-35"]:
            p["30-35"].append(df.loc[i]['location'])
            
    else:
        check["35-40"]=True
        if df.loc[i]['location'] not in p["35-40"]:
            p["35-40"].append(df.loc[i]['location'])
            
        
p
```

```python
age_location={}

for i in p.keys():
    for j in p[i]:
        age_location[i+" "+j]=[]

age_location
```

```python
product_df=pd.read_csv('final_final_product_dataset.csv')
product_df
```

```python
product_df=product_df[['product_id','ratings']]
product_df
```

```python
product_df=product_df.sort_values(by=['ratings'],ascending=False)
product_df
```

```python
new_df=df.merge(product_df,on='product_id')
new_df
```

```python
new_df=new_df.sort_values(by=['ratings'],ascending=False)
new_df
```

```python
for i in range(len(new_df)):
    if new_df.loc[i]['ratings'] != 'Get':
        key=""
        if new_df.loc[i]['age']<=20:
            key="18-20 "+new_df.loc[i]['location']

        elif new_df.loc[i]['age']<=25:
            key="20-25 "+new_df.loc[i]['location']

        elif new_df.loc[i]['age']<=30:
            key="25-30 "+new_df.loc[i]['location']

        elif new_df.loc[i]['age']<=35:
            key="30-35 "+new_df.loc[i]['location']

        else:
            key="35-40 "+new_df.loc[i]['location']
            
        if new_df.loc[i]['product_id'] not in age_location[key]:
            age_location[key].append(new_df.loc[i]['product_id'])
            
age_location
```

```python
for i in age_location.keys():
    age_location[i]=age_location[i][:100]

age_location
```

```python
import pickle
pickle.dump(age_location,open('age_location_recommend.pkl','wb'))
```
